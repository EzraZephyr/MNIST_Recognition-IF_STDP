```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'train.csv'
df = pd.read_csv(file_path)

best_train_acc = df.loc[df['Train_acc'].idxmax()]
best_test_acc = df.loc[df['Test_acc'].idxmax()]

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(df['Epoch'], df['Train_loss'], marker='o')
plt.title('Train Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Train Loss')

plt.subplot(3, 1, 2)
plt.plot(df['Epoch'], df['Train_acc'], marker='o')
plt.scatter(best_train_acc['Epoch'], best_train_acc['Train_acc'],
            color='red', s=80, label='Best Train Acc')
plt.annotate(
    f"{best_train_acc['Train_acc'] * 100:.2f}%",
    (best_train_acc['Epoch'], best_train_acc['Train_acc']),
    textcoords="offset points",
    xytext=(5,5),
    color='red'
)
plt.title('Train Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Train Accuracy')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df['Epoch'], df['Test_acc'], marker='o')
plt.scatter(best_test_acc['Epoch'], best_test_acc['Test_acc'],
            color='red', s=80, label='Best Test Acc')
plt.annotate(
    f"{best_test_acc['Test_acc'] * 100:.2f}%",
    (best_test_acc['Epoch'], best_test_acc['Test_acc']),
    textcoords="offset points",
    xytext=(5,5),
    color='red'
)
plt.title('Test Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Test Accuracy')
plt.legend()

plt.tight_layout()
plt.savefig('plot.png', dpi=300)
plt.show()
```
